```python
#data Cleaning
df = df.drop_duplicates(subset="species")
```

```python
sns.countplot(x='species', data=df)
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.show()

sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)
plt.show()
```

```python
sns.pairplot(df, hue='species', height=2.5)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0, 0].hist(df['sepal_length'], bins=3)
axes[0, 1].hist(df['sepal_width'], bins=8)
axes[1, 0].hist(df['petal_length'], bins=6)
axes[1, 1].hist(df['petal_width'], bins=5)
plt.show()
```

```python

plot = sns.FacetGrid(df, hue='species')
plot.map(sns.histplot, 'sepal_length').add_legend()
```

```python
def graph(y):
    sns.boxplot(x="species", y=y, data=df)

plt.figure(figsize=(12, 8))

plt.subplot(221)
graph('sepal_length')

plt.subplot(222)
graph('sepal_width')

plt.subplot(223)
graph('petal_length')

plt.subplot(224)
graph('petal_width')

plt.show()
```

```python
# Read the data
df = pd.read_csv('IRIS.csv')

# Calculate quartiles and interquartile range (IQR)
Q1 = np.percentile(df['sepal_length'], 25)
Q3 = np.percentile(df['sepal_length'], 75)
IQR = Q3 - Q1

# Identify and remove outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

df = df[(df['sepal_length'] >= lower_bound) & (df['sepal_length'] <= upper_bound)]

# Visualize the distribution after outlier removal
sns.boxplot(x=df['sepal_length'])
plt.show()
```
